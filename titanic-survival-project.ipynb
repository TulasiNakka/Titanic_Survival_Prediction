# Performing Feature Engineering


```python
train_data.isna().sum()
```

```python
train_data['PassengerId'].describe()
#train_data['Ticket'].describe()
```

```python
# Drop Cabin which is having high missing values and are irrelevant.
train_data = train_data.drop(columns=['Cabin'])

#drop passenger Id as it is not contributing for the survival predicition
train_data=train_data.drop(['PassengerId'],axis=1)

# drop ticket column
train_data =  train_data.drop(columns=['Ticket'])

# filling with 0 for NaN values
train_data.fillna(0, inplace=True)
```

```python
train_data.isna().sum()
```

```python
#filling the 0's with mean for Age 
age = train_data['Age'].mean()
print(age)
train_data['Age'] = train_data['Age'].fillna(age)
```

```python
train_data['Embarked'].describe()
```

```python
#As top is 'S' replacing the 2 0's substituted value with 'S'
top= 'S'
#embarked = train_data['Embarked'].value_counts()
train_data['Embarked'] = train_data['Embarked'].fillna(top)
```

```python
train_data['Embarked'].describe()
```

```python
test_data.isna().sum()
```

```python
# Drop Cabin which is having high missing values and are irrelevant.
test_data = test_data.drop(columns=['Cabin'])

# drop ticket column
test_data = test_data.drop(columns=['Ticket'])
# filling with 0 for NaN values
test_data.fillna(0, inplace=True)
```

```python
test_data.isna().sum()
```

```python
#filling the 0's with mean for Age 
age_test= test_data['Age'].mean()
print(age_test)
test_data['Age'] = test_data['Age'].fillna(age_test)

fare_test = test_data['Fare'].median()
test_data['Fare'] = test_data['Fare'].fillna(age_test)
#print(fare_test)
```

# Features 

```python
train_data.head(10)
```

## Labeling gender and Embarked into numerical data

```python
#Sex column data conversion
gender= {'male': 0, 'female': 1}
train_data['Sex'] = train_data['Sex'].map(gender)
test_data['Sex'] = test_data['Sex'].map(gender)

#Embarked column data conversion
embarked={'S': 0,'C': 1,'Q': 2}
train_data['Embarked'] = train_data['Embarked'].map(embarked)
#train_data['Embarked'] = train_data['Embarked'].astype(int)
test_data['Embarked'] = test_data['Embarked'].map(embarked)

print(train_data['Embarked'].unique())
```

```python
train_data.head(10)
```

# Random Forest Classifier - Accuracy is 0.78708 ~ 78.7%

```python

#version 7
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch", "Fare"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

# SVM Classifier - Accuracy is 0.66746 ~ 66.74%


```python

#import SVC classifier
from sklearn.svm import SVC

# import metrics to compute accuracy
from sklearn.metrics import accuracy_score

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch", "Fare","Age"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
# instantiate classifier with default hyperparameters
svc=SVC() 

# fit classifier to training set
svc.fit(X,y)
# make predictions on test set
predictions=svc.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)

print("Your submission was successfully saved!")
```

# XGB Boost - Accuracy is 0.76076 ~ 76.07%

```python

import xgboost as xgb
# import metrics to compute accuracy
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch", "Fare","Age","Embarked"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])


model = xgb.XGBClassifier(n_estimators = 30, 
                      learning_rate=0.1, 
                      max_depth= 10,
                      use_label_encoder=False, random_state =20)
model.fit(X, y)
predictions = model.predict(X_test)


output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)



print("Your submission was successfully saved!")
```
